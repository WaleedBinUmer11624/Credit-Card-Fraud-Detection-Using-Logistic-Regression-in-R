#**Class Imbalance**

```r
#generate a balanced data set named as fraud_data_TT
#perform under sampling on the fraud detection data set
fraud_Data_TT=ovun.sample(Class~.,data = Fraud_Detection, method = "under",
                                 p = 0.5,seed = 1)$data
```

#split data

```r
#split data
fraud_Training=sample(nrow(fraud_Data_TT),round(nrow(fraud_Data_TT)*0.75),replace = FALSE)
Credit_Data_Training=fraud_Data_TT[fraud_Training,];
Credit_Data_Test=fraud_Data_TT[fraud_Training,];
```

**Model#1**

```r
#fit a logistic regression model with class as dependent variiable and V1 as a predictor
fitmodel.lm1=glm(Class~V1, data = Credit_Data_Training, family = binomial)
#generte summary of the fitmodel.lm1
summary(fitmodel.lm1)
#create variables for the actual and prediicted variables
actual <- Credit_Data_Training$Class
predicted <- round(fitted(fitmodel.lm1))
#create coonfuusion matrix to compare the actual and predicted variables
xt <- xtabs(~ actual + predicted)
xt
#calcualte the accuracy of the model
accuracy <- sum(diag(xt))/sum(xt)
accuracy
#set the layout to display multiple plots
par(mfrow=c(2,2))
#create diagnostic plot for the logistic regression model
plot(fitmodel.lm1)
```

# **Predictions complete dataset**

```r
# Make predictions on the complete dataset
predicted_Credit_Data <- predict(fitmodel.lmfinal, newdata = Credit_Data_Training, type = "response")
# Threshold probabilities for binary classification
predicted_classes <- ifelse(predicted_Credit_Data > 0.5, 1, 0)
```

# **Confusion matrix**

```r
# Create confusion matrix for the complete dataset
conf_matrix <- table(Actual = Credit_Data_Training$Class, Predicted = predicted_classes)
```

```r
# Print confusion matrix and weighted accuracy
print("Confusion Matrix (Complete Dataset):")
print(conf_matrix)
```

# **weighted accuracy**

```r
# Calculate weighted accuracy
weighted_accuracy <- sum(diag(prop.table(conf_matrix, margin = 1))) / sum(prop.table(conf_matrix))
```
